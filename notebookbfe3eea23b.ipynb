```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

file_path = '/kaggle/input/top-movie-recommendation-dataset/movies.csv'
df = pd.read_csv(file_path)

fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off')  

table_data = df[['User_ID', 'Movie_ID', 'Movie_Name', 'Genre', 'Rating']]
table = ax.table(cellText=table_data.values,
                 colLabels=table_data.columns,
                 cellLoc='center',
                 loc='bottom')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)  

plt.suptitle('Raw Data Table', y=0.15, fontsize=16, ha='center')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/kaggle/input/top-movie-recommendation-dataset/movies.csv'
df = pd.read_csv(file_path)

grouped_data = df.groupby('Genre')[['Rating']].mean()

fig, ax = plt.subplots(figsize=(10, 6))
bar_plot = grouped_data.plot(kind='bar', ax=ax, rot=0)
plt.title('Average Rating of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.legend(['Movie Type Scores'])

ax.set_ylim(0, 5)

for p in bar_plot.patches:
    bar_plot.annotate(str(round(p.get_height(), 2)), (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```
